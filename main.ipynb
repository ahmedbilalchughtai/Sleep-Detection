```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
model = load_model(r'D:\Uni\6th Semester\AI\AI Project\models\model.keras')
```

```python
import cv2
import numpy as np
from pygame import mixer

mixer.init()
sound = mixer.Sound(r'D:\Uni\6th Semester\AI\AI Project\alarm.wav')

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640/2)  # Set frame width
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480/2)  # Set frame height
cap.set(cv2.CAP_PROP_FPS, 25)  # Set frames per second

Score = 0
alarm_playing = False
frame_count = 0  # To process every nth frame

face_orientation = "straight"  # To track face orientation

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)  # Flip the frame horizontally
    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Detect eyes within the face region
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyes:
            # Preprocess eye
            eye = roi_color[ey:ey + eh, ex:ex + ew]
            eye = cv2.resize(eye, (80, 80))
            eye = eye / 255.0
            eye = eye.reshape(80, 80, 3)
            eye = np.expand_dims(eye, axis=0)

            # Model prediction
            prediction = model.predict(eye)

            # If eyes are closed
            if prediction[0][0] > 0.30:
                Score += 1
                print("sleep")
            else:
                Score = 0

            # Play alarm if eyes are closed for a sustained period
            if Score > 5:
                if not alarm_playing:
                    sound.play(-1)  # Play alarm indefinitely
                    alarm_playing = True
            else:
                if alarm_playing and Score <= 0:
                    sound.stop()
                    alarm_playing = False

        # Detect face orientation based on face dimensions
        if h < y + h * 0.6:  # Face tilted downward
            face_orientation = "downward"
            if not alarm_playing:
                sound.play(-1)  # Play alarm indefinitely
                alarm_playing = True
        else:
            face_orientation = "straight"
            if alarm_playing and Score <= 0:
                sound.stop()
                alarm_playing = False

    cv2.putText(frame, 'Score: ' + str(Score), (10, height - 20), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=1, color=(255, 255, 255),
                thickness=1, lineType=cv2.LINE_AA)
    cv2.putText(frame, 'Face Orientation: ' + face_orientation, (10, height - 40), fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL, fontScale=1, color=(255, 255, 255),
                thickness=1, lineType=cv2.LINE_AA)

    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Use a smaller waitKey value for smoother frame rate
        break

cap.release()
cv2.destroyAllWindows()
```
